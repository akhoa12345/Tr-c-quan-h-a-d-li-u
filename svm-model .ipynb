```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/kaggle/input/weather2/DataForFinal.xlsx")
df
```

```python
transformed_data = pd.get_dummies(df[["YEAR", "MO", "DY", "QV2M", "RH2M", "PRECTOTCORR", "PS", "WS2M", "T2M_RANGE", "T2M", "WS10M_MAX", "WS10M_MIN", "WS10M", "WS10M_RANGE", "WS50M", "WS50M_MAX", "WS50M_MIN", "WS50M_RANGE", "TS" ]])
transformed_data
```

```python
x_data = transformed_data.values
y_data = df["City"].values
```

```python
from sklearn.model_selection import train_test_split
x_data_train, x_data_test, y_data_train, y_data_test = train_test_split(x_data, y_data,test_size=0.2,random_state=1 )
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(x_data_train, y_data_train)

#Predict the response for test dataset
y_pred = clf.predict(x_data_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_data_test, y_pred))
```
