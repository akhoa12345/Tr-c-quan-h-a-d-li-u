```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/kaggle/input/weather1/DataForFinal.xlsx")
df
```

```python
transformed_data = pd.get_dummies(df[["YEAR", "MO", "DY", "QV2M", "RH2M", "PRECTOTCORR", "PS", "WS2M", "T2M_RANGE", "WS10M_MAX", "WS10M_MIN", "WS10M", "WS10M_RANGE", "WS50M", "WS50M_MAX", "WS50M_MIN", "WS50M_RANGE", "TS", "City" ]])
transformed_data
```

```python
x_data = transformed_data.values
y_data = df["T2M"].values
```

```python
from sklearn.model_selection import train_test_split
x_data_train, x_data_test, y_data_train, y_data_test = train_test_split(x_data, y_data,test_size=0.2,random_state=1 )
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(x_data_train, y_data_train)

Y_pred = lm.predict(x_data_test)

plt.scatter(y_data_test, Y_pred)
plt.xlabel("Value: $Y_i$")
plt.ylabel("Predicted Value: $\hat{Y}_i$")
plt.title("Value vs Value prices: $Y_i$ vs $\hat{Y}_i$")
```
